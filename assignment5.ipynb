```python
import pandas as pd
```

```python
df=pd.read_csv("C:\\Users\\Lenovo\\Documents\\ML Club\\PCE-CS-III-ML-Club\\Data Sets\\Assignment5.csv",names=['emp','senior'])
df
```

```python
my_dict=dict(zip(df.emp,df.senior))
my_dict
```

```python
result = []
for key in my_dict :
    temp = []
    temp.append(key)
    tempkey = key;
    while tempkey != my_dict[tempkey] :
        tempkey = my_dict[tempkey]
        temp.append(tempkey)
    result.append(temp)
result
```

```python
dataframe=pd.DataFrame(result)
```
